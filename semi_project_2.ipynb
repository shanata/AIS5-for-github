```python
data_df = pd.read_csv('semi_2_data/titanic.csv')
data_df
```

## Feature Engineering

```python
X_data['Name'] = X_data['Name'].str.split(',').apply(lambda x:x[1])
X_data['Name'] = X_data['Name'].str.split('.').apply(lambda x:x[0])
X_data['Sex'] = X_data['Sex'].replace({'male':0, 'female':1})
X_data['Fare'] = np.log1p(X_data['Fare'])
X_data['Cabin'] = X_data['Cabin'].str[0]
X_data['Cabin'] = X_data['Cabin'].fillna('Z')
X_data['Cabin'] = X_data['Cabin'].map({'A':0, 'B':1, 'C':2, 'D':3, 'E':4,'F':5,'G':6, 'T':7,'Z':8})
X_data['Embarked'] = X_data['Embarked'].fillna('S')
X_data['Embarked'] = X_data['Embarked'].map({'C':0, 'Q':1, 'S':2})
X_data
```

```python
# Age
# train_test_split
age_train = X_data[~X_data['Age'].isnull()]
age_test = X_data[X_data['Age'].isnull()]

age_X_train = age_train.drop(['Age'], axis=1)
age_y_train = age_train[['Age']]

age_X_test = age_test.drop(['Age'], axis=1)
age_y_test = age_test[['Age']]
```

```python
age_X_train
```

```python
#태용수정시작
x_train, x_test, y_train, y_test = model_selection.train_test_split(age_X_train, age_y_train, test_size=0.3, random_state=0)
```

```python
# Make Pipeline for feature-transformer (StandardScaler & OneHotEncoder)

numeric_features = ['Sex','SibSp','Parch','Fare']
numeric_transformer = StandardScaler()

categorical_features = ['Pclass','Embarked']
categorical_transformer = OneHotEncoder(categories='auto', handle_unknown='ignore') 

preprocessor = ColumnTransformer(
    transformers=[('num', numeric_transformer, numeric_features),
                  ('cat', categorical_transformer, categorical_features)])
```

```python
# Pipeline usage - 전처리용 (fit & transform)
preprocessor_pipe = Pipeline(steps=[('preprocessor', preprocessor)])
preprocessor_pipe.fit(x_train)

x_train_transformed = preprocessor_pipe.transform(x_train)
x_test_transformed = preprocessor_pipe.transform(x_test)
age_X_test_transformed = preprocessor_pipe.transform(age_X_test)
```

```python
# LinearRegression (선형회귀) 모델 객체 만들기
# model = LinearRegression()
# model =GradientBoostingRegressor()
model =XGBRegressor()
# model =LGBMRegressor()

# 모델 객체 학습시키기
model.fit(x_train_transformed, y_train)

# 학습이 끝난 모델 테스트하기
print('MSE: ', mean_squared_error(model.predict(x_train_transformed), y_train))
print('MSE: ', mean_squared_error(model.predict(x_test_transformed), y_test))
print("========================")
```

```python
#결측값 예측하기
predicted_y = model.predict(age_X_test_transformed)
```

```python
# null 값이었던 Age 데이터에 예측된 값 넣어주기
age_y_test['Age'] = predicted_y
age_y_test
```

```python
# 데이터 합치기 #태용 수정 끝
filled_X = pd.concat([age_X_train, age_X_test], axis=0)
filled_y = pd.concat([age_y_train, age_y_test], axis=0)

filled_X_data = pd.concat([filled_X, filled_y], axis=1, sort=True)
X_data = filled_X_data
X_data.head(3)
```

```python
X_data = np.array(X_data)
y_data = np.array(y_data)

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.3, random_state=13)

numeric_features = [5,8]
numeric_transformer = RobustScaler()
categorical_features = [0,1,2,3,4,6,7]
categorical_transformer = OneHotEncoder(categories='auto', handle_unknown='ignore') 

preprocessor = ColumnTransformer(
    transformers=[('num', numeric_transformer, numeric_features),
                  ('cat', categorical_transformer, categorical_features)])
preprocessor_pipe = Pipeline(steps=[('preprocessor', preprocessor)])
preprocessor_pipe.fit(X_train)

X_train_transformed = preprocessor_pipe.transform(X_train)
X_test_transformed = preprocessor_pipe.transform(X_test)
```

## Modeling

```python
def modeling_metrics(y_test, pred):
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    f1 = f1_score(y_test, pred)

    print('Accuracy(정확도):', accuracy)
    print('Precision(정밀도):', precision)
    print('Recall(재현율):', recall)
    print('F1:', f1)

    return [accuracy, precision, recall, f1]

def modeling_roc(y_test,pred):
    fpr, tpr, _ = roc_curve(y_true=y_test,y_score=pred[:,1])
    roc_auc = auc(fpr, tpr)
    
    print('ROC_AUC:',roc_auc)
    
    return [fpr,tpr,roc_auc]
```

```python
params_grid = {'SVM': {'C' : [0.1, 1, 10, 100], 
                       'gamma' : [1, 0.1, 0.01, 0.001, 0.0001],
                       'kernel' : ['rbf']},
               'kNN': {'n_neighbors' : [3, 5, 10, 20],
                       'leaf_size' : [20, 30, 40],
                       'p' : [1, 2]},
               'LR' : {'C': [0.1, 1, 10, 100],
                       'max_iter' : [100, 200]},
               'DT': {'criterion' : ['gini', 'entropy'],
                      'max_depth' : [0, 1, 2, 3, 4, 5], 
                      'min_samples_split' : [1, 2, 3], 
                      'min_samples_leaf' : [1, 2, 3]},
               'RF': {'criterion' : ['gini', 'entropy'],
                      'n_estimators' : [50, 100, 150],
                      'min_samples_split' : [1, 2, 3], 
                      'min_samples_leaf' : [1, 2, 3]},               
               'GB' : {'learning_rate' : [1, 0.1, 0.01, 0.001],
                       'n_estimators' : [50, 100, 150],
                       'min_samples_split' : [1, 2, 3],
                       'min_samples_leaf' : [1, 2, 3]},
               'ExtraTree': {'max_depth' : [0, 1, 2, 3, 4, 5],
                             'n_estimators' : [50, 100, 150],
                             'min_samples_split' : [1, 2, 3],
                             'min_samples_leaf' : [1, 2, 3]},
               'XGB' : {'max_depth' : [-1, 0, 1, 2, 3, 4, 5],
                        'n_estimators' : [50, 100, 150],
                        'learning_rate' : [1, 0.1, 0.01, 0.001],},
               'LightGBM' : {'max_depth' : [-1, 0, 1, 2, 3, 4, 5],
                             'n_estimators' : [50, 100, 150],
                             'learning_rate' : [1, 0.1, 0.01, 0.001]}}

model_list = {'SVM': SVC(probability=True),
              'kNN': neighbors.KNeighborsClassifier(),
              'LR': linear_model.LogisticRegression(),
              'DT': tree.DecisionTreeClassifier(),
              'RF': ensemble.RandomForestClassifier(),
              'GB': ensemble.GradientBoostingClassifier(),
              'ExtraTree': ensemble.ExtraTreesClassifier(),
              'XGB': XGBClassifier(),
              'LightGBM': LGBMClassifier()}

grid_metrics_result={}
for model_name, model in model_list.items():
    grid_model = GridSearchCV(model, params_grid[model_name], refit=True, verbose=1)
    grid_model.fit(X_train_transformed, y_train)
    print(model_name)
    print('The best parameters are ', grid_model.best_params_)
    
    pred = grid_model.predict(X_test_transformed)
    pred_proba = grid_model.predict_proba(X_test_transformed)[:,1]
    
    grid_metrics_result[model_name] = modeling_metrics(y_test, pred)
    print("========================")
    
grid_metrics_df = pd.DataFrame.from_dict(grid_metrics_result)
grid_metrics_df.index = ['Accuracy', 'Precision', 'Recall', 'F1']
grid_metrics_df
```

```python
from vecstack import StackingTransformer
```

```python
estimators = [('XGB', XGBClassifier(learning_rate=0.1, max_depth=3, n_estimators=100)),
              ('LightGBM', LGBMClassifier(learning_rate=0.1, max_depth=3, n_estimators=150)),
              ('DT', tree.DecisionTreeClassifier(max_depth=4, min_samples_leaf=1, min_samples_split=3)),]
    
# Initialize StackingTransformer
# x,y 안들어가고, 마치 모델 객체 생성하듯이
stack = StackingTransformer(estimators, 
                            regression = False, 
                            metric = accuracy_score, 
                            n_folds = 5, stratified = True, shuffle = True, 
                            random_state = 0, verbose = 0) 
                            
# Fit
stack = stack.fit(X_train_transformed, y_train)

# Get your stacked features
S_train = stack.transform(X_train_transformed)
S_test = stack.transform(X_test_transformed)

# Use 2nd level estimator with stacked features
model = ensemble.RandomForestClassifier(min_samples_leaf=2, min_samples_split=3, n_estimators=100)
model = model.fit(S_train, y_train) 
y_pred = model.predict(S_test) 
print('Final prediction score: [%.8f]' % accuracy_score(y_test, y_pred))
```

```python
estimators = [('XGB', XGBClassifier(learning_rate=0.1, max_depth=3, n_estimators=100)),
              ('LightGBM', LGBMClassifier(learning_rate=0.1, max_depth=3, n_estimators=150)),
              ('DT', tree.DecisionTreeClassifier(max_depth=4, min_samples_leaf=1, min_samples_split=3)),]
    
# Initialize StackingTransformer
# x,y 안들어가고, 마치 모델 객체 생성하듯이
stack = StackingTransformer(estimators, 
                            regression = False, 
                            metric = accuracy_score, 
                            n_folds = 5, stratified = True, shuffle = True, 
                            random_state = 0, verbose = 0) 
                            
# Fit
stack = stack.fit(X_train_transformed, y_train)

# Get your stacked features
S_train = stack.transform(X_train_transformed)
S_test = stack.transform(X_test_transformed)

# Use 2nd level estimator with stacked features
model = ensemble.GradientBoostingClassifier(learning_rate=0.01, min_samples_leaf=2, min_samples_split=2, n_estimators=150)
model = model.fit(S_train, y_train) 
y_pred = model.predict(S_test) 
print('Final prediction score: [%.8f]' % accuracy_score(y_test, y_pred))
```
